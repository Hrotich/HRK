```python
# Let's import the pandas library 
# 
import pandas as pd
```

```python
# loading our cells_geo csv dataset
url = "https://raw.githubusercontent.com/Hrotich/week3001/main/cells_geo.csv"

df = pd.read_csv(url,  delimiter = ';')
df.head(8)
```

```python
# to find missing values
df.isnull()
```
